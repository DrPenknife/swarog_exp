```python
import pandas as pd
import sqlite3
from IPython.display import display

con = sqlite3.connect("swarog_wyniki_v1.sqlite")
df = pd.read_sql_query("""
select dataset, method, datetime(date, 'unixepoch') as T, 
(case when avg(value) > 0.75 then "YES" else "  " end ) as "B.A. > 0.75",
metric, avg(value), count(*) as N,
max(value) - min(value) as "max-min", 
sqrt(AVG(value*value) - AVG(value)*AVG(value)) as "std."
from wyniki where metric like 'Bal%' group by dataset,date,method,metric order by dataset,date asc;
""", con)

display(df)
con.close()
```
